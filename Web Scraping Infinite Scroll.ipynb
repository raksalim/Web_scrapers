```python
from selenium import webdriver
import sys
import unittest, time, re

import re
import requests
from bs4 import BeautifulSoup
from os.path  import basename
import wget



SCROLL_PAUSE_TIME = 5
```

```python
url = input("Enter URL here: ")
```

```python

driver = webdriver.Chrome('C:/Users/raksa/chromedriver.exe')
driver.get(URL)
```

```python

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

posts = driver.find_elements_by_class_name("rg_i Q4LuWd")

for block in posts:
    print(block)
```

```python
soup = BeautifulSoup(driver.page_source, "lxml")
```

```python
img_div = soup.find_all("img", class_="rg_i Q4LuWd")
```

```python
#list of img links
img_links = []

for i in img_div:
    img_links.append(i.get("src"))
    
#print(img_links)
```

```python
#download imiage from list of [names: source]
name = input("file name:? ")
directory_name = input("Input folder name: ")
directory_name = directory_name.replace("\\", "/")
for i in range(len(img_links)):
    img_url = img_links[i]
    file_name =  "/"+name+str(i)+".jpg"
    file_path = directory_name+file_name

    try:

        img_file = wget.download(img_url, file_path)
        print("downloaded: ", file_name)
    except:
        print("did not download")
        pass
    
    print("Done", img_links[i], file_name)
```
