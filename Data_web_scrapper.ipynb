```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# enter URL here

url = input("Enter URL here: ")
```

```python
# Connect and request webpage

response = requests.get(url)
print(response.status_code)

html = response.text
```

```python
print(html)
```

```python
# Create Soup Object

soup = BeautifulSoup(response.content, "lxml")
```

```python
# list to store results
textContent = []

for i in soup.find_all("tr"):
    textContent.append([])
    for row in i:
        textContent[len(textContent)-1].append(row.text)

print(textContent)



```

```python
text_content_df = pd.DataFrame(textContent)

print(text_content_df.head)
```

```python
file_name = input("Enter fiel name: ")
file_name = file_name + ".csv"

text_content_df.to_csv(file_name)
```
